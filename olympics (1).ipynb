```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#load dataset
athletes=pd.read_csv("athlete_events.csv")
region=pd.read_csv("noc_regions.csv")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Filter the DataFrame for India
india_medals = athletes_df[athletes_df['Region'] == 'India']

# Count the number of gold medals
gold_medals_count = india_medals[india_medals['Medal'] == 'Gold'].shape[0]

# Create a pie chart to visualize the number of gold medals
labels = ['Gold Medals', 'No Medals']
sizes = [gold_medals_count, india_medals.shape[0] - gold_medals_count]
colors = ['gold', 'lightgrey']
explode = (0.1, 0)  # explode the 1st slice (Gold Medals)

plt.figure(figsize=(8, 8))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140, shadow=True)
plt.title('Distribution of Gold Medals for India')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Display the pie chart
plt.show()
```
